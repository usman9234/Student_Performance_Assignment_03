# Task 2: Descriptive Statistics

In this task, descriptive statistics are calculated for math score, reading score, and writing score. The analysis includes mean, median, mode, standard deviation, minimum, maximum, mean vs median comparison, groupby analysis, and observations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/StudentsPerformance.csv")
df.head()
```

```python
score_columns = ['math score', 'reading score', 'writing score']

df[score_columns].head()
```

```python
mean_values = df[score_columns].mean()
mean_values
```

```python
median_values = df[score_columns].median()
median_values
```

```python
mode_values = df[score_columns].mode().iloc[0]
mode_values
```

```python
std_values = df[score_columns].std()
std_values
```

```python
min_values = df[score_columns].min()
min_values
```

```python
max_values = df[score_columns].max()
max_values
```

```python
statistics_table = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Mode': mode_values,
    'Standard Deviation': std_values,
    'Minimum': min_values,
    'Maximum': max_values
})

statistics_table
```

```python
gender_group = df.groupby('gender')[score_columns].mean()
gender_group
```

```python
test_prep_group = df.groupby('test preparation course')[score_columns].mean()
test_prep_group
```

```python
lunch_group = df.groupby('lunch')[score_columns].mean()
lunch_group
```

# Task 3: Data Cleaning

In this task, a messy version of the dataset is created to demonstrate data cleaning. The cleaning process includes identifying missing values, duplicate rows, and outliers. Missing values are handled, duplicates are removed, invalid score values are cleaned, and the final cleaned dataset is saved.

```python
df = pd.read_csv("../data/StudentsPerformance.csv")
df.head()
```

```python
messy_df = df.copy()

# Adding missing values
messy_df.loc[0, 'math score'] = np.nan
messy_df.loc[1, 'reading score'] = np.nan
messy_df.loc[2, 'writing score'] = np.nan

# Adding duplicate rows
messy_df = pd.concat([messy_df, messy_df.iloc[0:5]], ignore_index=True)

# Adding outliers / invalid values
messy_df.loc[10, 'math score'] = 150
messy_df.loc[11, 'reading score'] = -20
messy_df.loc[12, 'writing score'] = 130

messy_df.head(15)
```

```python
missing_values_before = messy_df.isnull().sum()
missing_values_before
```

```python
duplicates_before = messy_df.duplicated().sum()
duplicates_before
```

```python
messy_df[['math score', 'reading score', 'writing score']].describe()
```

```python
before_rows = messy_df.shape[0]
before_missing = messy_df.isnull().sum().sum()
before_duplicates = messy_df.duplicated().sum()

print("Rows before cleaning:", before_rows)
print("Missing values before cleaning:", before_missing)
print("Duplicate rows before cleaning:", before_duplicates)
```

```python
messy_df['math score'] = messy_df['math score'].fillna(messy_df['math score'].median())
messy_df['reading score'] = messy_df['reading score'].fillna(messy_df['reading score'].median())
messy_df['writing score'] = messy_df['writing score'].fillna(messy_df['writing score'].median())

messy_df.isnull().sum()
```

```python
messy_df = messy_df.drop_duplicates()

messy_df.duplicated().sum()
```

```python
cleaned_df = messy_df[
    (messy_df['math score'] >= 0) & (messy_df['math score'] <= 100) &
    (messy_df['reading score'] >= 0) & (messy_df['reading score'] <= 100) &
    (messy_df['writing score'] >= 0) & (messy_df['writing score'] <= 100)
]

cleaned_df[['math score', 'reading score', 'writing score']].describe()
```

```python
after_rows = cleaned_df.shape[0]
after_missing = cleaned_df.isnull().sum().sum()
after_duplicates = cleaned_df.duplicated().sum()

print("Rows after cleaning:", after_rows)
print("Missing values after cleaning:", after_missing)
print("Duplicate rows after cleaning:", after_duplicates)
```

```python
cleaning_comparison = pd.DataFrame({
    'Before Cleaning': [before_rows, before_missing, before_duplicates],
    'After Cleaning': [after_rows, after_missing, after_duplicates]
}, index=['Total Rows', 'Missing Values', 'Duplicate Rows'])

cleaning_comparison
```

```python
cleaned_df.to_csv("../data/Cleaned_StudentsPerformance.csv", index=False)

print("Cleaned dataset saved successfully.")
```
